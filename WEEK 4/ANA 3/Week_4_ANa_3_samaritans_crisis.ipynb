```python
years = list(range(2001, 2025))

values = [
4914,4951,5005,4960,4713,4482,4505,4872,
4599,4664,4782,5207,5056,4872,5055,5006,
5235,5780,5873,5797,5819,5846,5418,2395
]

trend_df = pd.DataFrame({
    "Year": years,
    "Occurrences": values
})

trend_df.head()
```

```python
contacts = pd.read_csv("MHB_2425_Chapter 6.csv")

contacts.shape
```

```python
contacts.columns.tolist()
```

```python
contacts["REPORTING_PERIOD_START"].nunique()
```

```python
contacts["REPORTING_PERIOD_START"].unique()[:10]
```

```python
contacts["LEVEL_TWO_DESCRIPTION"].dropna().unique()
```

```python
crisis_services = [
    "Crisis Resolution Team/Home Treatment Service",
    "24/7 Crisis Response Line",
    "Crisis Cafè/Safe Haven/Sanctuary Service",
    "Walk-in Crisis Assessment Unit Service",
    "Crisis House Service"
]

call_volume = contacts[
    contacts["LEVEL_TWO_DESCRIPTION"].isin(crisis_services)
]

call_volume[[
    "LEVEL_TWO_DESCRIPTION",
    "METRIC_VALUE"
]].head(20)
```

```python
call_summary = (
    call_volume
    .groupby("LEVEL_TWO_DESCRIPTION")["METRIC_VALUE"]
    .sum()
    .reset_index()
    .sort_values("METRIC_VALUE", ascending=False)
)

call_summary
```

```python
plt.figure(figsize=(59,26))

plt.bar(
    call_summary["LEVEL_TWO_DESCRIPTION"],
    call_summary["METRIC_VALUE"]
)

plt.title("Crisis Service Contact Volumes")
plt.xlabel("Crisis Service")
plt.ylabel("Total Contacts")

plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
call_summary
```

```python
call_volume["METRIC_VALUE"] = pd.to_numeric(
    call_volume["METRIC_VALUE"],
    errors="coerce"
)

call_summary = (
    call_volume
    .groupby("LEVEL_TWO_DESCRIPTION")["METRIC_VALUE"]
    .sum()
    .reset_index()
    .sort_values("METRIC_VALUE", ascending=False)
)

call_summary
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.barh(
    call_summary["LEVEL_TWO_DESCRIPTION"],
    call_summary["METRIC_VALUE"] / 1000
)

plt.title("Crisis Service Contact Volumes (Thousands)")
plt.xlabel("Contacts (Thousands)")
plt.ylabel("Service")

plt.tight_layout()
plt.show()
```
