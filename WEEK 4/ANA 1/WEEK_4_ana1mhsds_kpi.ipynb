```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    "MHSDS Time_Series_data_Feb_Perf_2026.csv"
)

df.head()
```

contacts kpi


```python
contacts_df = df[

    df["MEASURE_NAME"].str.contains(
        "contact",
        case=False,
        na=False
    )

]

contacts_df.head(10)
```

admission kpi

```python
admissions_df = df[

    df["MEASURE_NAME"].str.contains(

        "inpatient|admitted|hospital",

        case=False,
        na=False

    )

]

admissions_df.head(10)
```

bed kpi

```python
beds_df = df[

    df["MEASURE_NAME"].str.contains(

        "bed|ward",

        case=False,
        na=False

    )

]

beds_df.head(10)
```

rtt kpi

```python
rtt_df = df[

    df["MEASURE_NAME"].str.contains(

        "waiting|RTT",

        case=False,
        na=False

    )

]

rtt_df.head(10)
```

crisis call/ referrals kpi

```python
crisis_df = df[

    df["MEASURE_NAME"].str.contains(

        "referral|crisis",

        case=False,
        na=False

    )

]

crisis_df.head(10)
```

kpi summary table

```python
summary = {

    "Contacts KPI":
    pd.to_numeric(
        contacts_df["MEASURE_VALUE"],
        errors="coerce"
    ).sum(),

    "Admissions KPI":
    pd.to_numeric(
        admissions_df["MEASURE_VALUE"],
        errors="coerce"
    ).sum(),

    "Beds KPI":
    pd.to_numeric(
        beds_df["MEASURE_VALUE"],
        errors="coerce"
    ).sum(),

    "RTT KPI":
    pd.to_numeric(
        rtt_df["MEASURE_VALUE"],
        errors="coerce"
    ).sum(),

    "Crisis KPI":
    pd.to_numeric(
        crisis_df["MEASURE_VALUE"],
        errors="coerce"
    ).sum()

}

summary_df = pd.DataFrame(

    list(summary.items()),

    columns=[

        "KPI Category",
        "Total Value"

    ]

)

summary_df
```

```python
summary_df["Total Value"] = summary_df["Total Value"].astype(int)

summary_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.bar(

    summary_df["KPI Category"],

    summary_df["Total Value"]

)

plt.title(
    "NHS Mental Health KPI Analysis"
)

plt.xlabel(
    "KPI Categories"
)

plt.ylabel(
    "Total KPI Values"
)

plt.xticks(rotation=10)

plt.show()
```

```python
source_mapping = pd.DataFrame({

    "KPI Category": [

        "Contacts KPI",
        "Admissions KPI",
        "Beds KPI",
        "RTT KPI",
        "Crisis KPI"

    ],

    "Source Column": [

        "MEASURE_NAME",
        "MEASURE_NAME",
        "MEASURE_NAME",
        "MEASURE_NAME",
        "MEASURE_NAME"

    ],

    "Keyword Filter": [

        "contact",
        "inpatient|admitted|hospital",
        "bed|ward",
        "waiting|RTT",
        "referral|crisis"

    ]

})

source_mapping
```

```python
summary_df.to_csv(

    "ANA1_KPI_Summary.csv",

    index=False

)

print(
    "KPI Summary CSV Saved"
)
```

```python
source_mapping.to_csv(

    "ANA1_Source_Mapping.csv",

    index=False

)

print(
    "Source Mapping CSV Saved"
)
```

key findings:

1.Contacts KPI recorded the highest value among all NHS mental health indicators.

2. Crisis-related measures also showed significantly high values, indicating increased demand for urgent mental health support services.

3. Admissions KPI recorded comparatively lower values than contacts and crisis indicators.

4. Bed-related metrics highlighted substantial inpatient mental health service utilization.

5. RTT (Referral to Treatment Time) measures indicate ongoing waiting-time pressures within mental health services.
