```python
# APMS vs CSEW Gender Comparison

categories = ["Men", "Women"]

# APMS values
apms_values = [9.6, 16.5]

# CSEW values
csew_values = [6.1, 11.2]

x = np.arange(len(categories))
width = 0.35

plt.figure(figsize=(8,5))

plt.bar(
    x - width/2,
    apms_values,
    width,
    label="APMS: Any Treatment"
)

plt.bar(
    x + width/2,
    csew_values,
    width,
    label="CSEW: Any Domestic Abuse"
)

plt.xticks(x, categories)

plt.ylabel("Percentage (%)")

plt.xlabel("Gender")

plt.title(
    "APMS vs CSEW Gender Comparison"
)

plt.legend()

plt.grid(axis='y')

plt.show()
```

trend comparison

```python
# APMS vs CSEW Trend Comparison

# APMS values
apms_age = [
    6.1,
    11.7,
    13.8,
    15.9,
    17.6,
    13.1,
    13.2
]

apms_groups = [
    "16-24",
    "25-34",
    "35-44",
    "45-54",
    "55-64",
    "65-74",
    "75+"
]

# CSEW values
csew_values = [
    6.06,
    5.72,
    5.13,
    5.35
]

csew_years = [
    "2019-20",
    "2021-22",
    "2022-23",
    "2023-24"
]

plt.figure(figsize=(10,5))

plt.plot(
    apms_groups,
    apms_age,
    marker='o',
    linewidth=2,
    label="APMS Treatment Trend"
)

plt.plot(
    csew_years,
    csew_values,
    marker='o',
    linewidth=2,
    label="CSEW Abuse Trend"
)

plt.title(
    "APMS vs CSEW Trend Comparison"
)

plt.ylabel("Percentage (%)")

plt.legend()

plt.grid()

plt.show()
```
