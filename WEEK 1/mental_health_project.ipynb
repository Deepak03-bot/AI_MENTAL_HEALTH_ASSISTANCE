```python
# Clean dubara load karo (fresh start best hai)
df = pd.read_excel("chapter 2.xls", sheet_name="2.1", skiprows=5)

df = df.rename(columns={df.columns[0]: "Category"})
df = df[df["Category"].notna()].reset_index(drop=True)

df
```

```python
df_men = df.iloc[0:6]
df_men
```

```python
df_women = df.iloc[7:13]
df_women
```

```python
# Numeric convert karo (safe)
df_men_num = df_men.set_index("Category").apply(pd.to_numeric, errors='coerce')
df_women_num = df_women.set_index("Category").apply(pd.to_numeric, errors='coerce')

# Average nikal lo (simple representation)
men_avg = df_men_num.mean(axis=1)
women_avg = df_women_num.mean(axis=1)

# Combine
df_compare = pd.DataFrame({
    "Men": men_avg,
    "Women": women_avg
})

df_compare
```

```python
import matplotlib.pyplot as plt
df_compare.plot(kind='bar', figsize=(10,5))

plt.title("Age-wise Mental Health Severity Comparison")
plt.xlabel("Age / Severity Categories")
plt.ylabel("Percentage")
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.grid(axis='y')
plt.show()
```

```python
# All adults ka index find karo
idx_all = df[df["Category"] == "All adults"].index[0]

# Uske neeche ke 6 rows = All adults data
df_all = df.iloc[idx_all+1 : idx_all+7]

df_all
```

```python
# Convert numeric
df_all_num = df_all.set_index("Category").apply(pd.to_numeric, errors='coerce')

# Average
all_avg = df_all_num.mean(axis=1)

# Combine all 3
df_compare_all = pd.DataFrame({
    "Men": men_avg,
    "Women": women_avg,
    "All Adults": all_avg
})

df_compare_all
```

```python
import matplotlib.pyplot as plt

df_compare_all.plot(kind='bar', figsize=(10,5))

plt.title("Mental Health Severity Comparison (Men vs Women vs All Adults)")
plt.ylabel("Percentage")
plt.xlabel("Categories")
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y')

plt.show()
```

```python
!pip install geopandas

import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Dataset
region_data = pd.DataFrame({
    "Region": [
        "North East",
        "North West",
        "Yorkshire and The Humber",
        "East Midlands",
        "West Midlands",
        "East of England",
        "London",
        "South East",
        "South West"
    ],

    "Any_CMD": [
        15.7,
        19.0,
        16.7,
        16.7,
        18.4,
        14.4,
        18.0,
        13.6,
        20.9
    ]
})

# UK map directly load
url = "https://raw.githubusercontent.com/martinjc/UK-GeoJSON/master/json/administrative/eng/lad.json"

uk_map = gpd.read_file(url)

# Plot base map
fig, ax = plt.subplots(figsize=(10,10))

uk_map.plot(ax=ax, color='lightgrey', edgecolor='black')

plt.title("UK Regional Mental Health Map")
plt.show()
```

```python
# Region-wise approximate coordinates
region_points = {
    "North East": (54.9783, -1.6178),
    "North West": (53.4808, -2.2426),
    "Yorkshire and The Humber": (53.8008, -1.5491),
    "East Midlands": (52.9548, -1.1581),
    "West Midlands": (52.4862, -1.8904),
    "East of England": (52.2405, 0.9027),
    "London": (51.5074, -0.1278),
    "South East": (51.4545, -0.9781),
    "South West": (50.7184, -3.5339)
}

# Plot base map again
fig, ax = plt.subplots(figsize=(10,10))

uk_map.plot(ax=ax, color='lightgrey', edgecolor='black')

# Add colored points
for i, row in region_data.iterrows():

    lat, lon = region_points[row["Region"]]

    ax.scatter(
        lon,
        lat,
        s=row["Any_CMD"] * 30,   # size based on value
        alpha=0.7,
        label=row["Region"]
    )

# Title
plt.title("UK Regional Mental Health Disorder Prevalence")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))

# Base UK map
uk_map.plot(ax=ax, color='lightgrey', edgecolor='black')

# Scatter with intensity colors
scatter = ax.scatter(
    [region_points[r][1] for r in region_data["Region"]],  # longitude
    [region_points[r][0] for r in region_data["Region"]],  # latitude
    s=region_data["Any_CMD"] * 40,
    c=region_data["Any_CMD"],
    cmap='Reds',
    alpha=0.8
)

# Color bar
cbar = plt.colorbar(scatter)
cbar.set_label("Mental Health Disorder Rate (%)")

# Title
plt.title("UK Regional Mental Health Disorder Intensity Map")

plt.show()
```
